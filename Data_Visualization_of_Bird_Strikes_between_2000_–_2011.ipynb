```python
import pandas as pd

# Read the Excel file
bird_strikes_data = pd.read_excel("/content/Bird Strikes data.xlsx")

# Visuals Depicting the Number of Bird Strikes
import matplotlib.pyplot as plt

# Yearly Analysis & Bird Strikes in the US
bird_strikes_data['Year'] = bird_strikes_data['FlightDate'].dt.year
yearly_bird_strikes = bird_strikes_data.groupby('Year').size()

# Plotting Yearly Analysis
plt.figure(figsize=(10, 6))
yearly_bird_strikes.plot(kind='bar', color='skyblue')
plt.title('Yearly Bird Strikes')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Top 10 US Airlines in terms of having encountered bird strikes
top_10_airlines = bird_strikes_data['Aircraft: Airline/Operator'].value_counts().nlargest(10)

# Plotting Top 10 Airlines
plt.figure(figsize=(10, 6))
top_10_airlines.plot(kind='bar', color='orange')
plt.title('Top 10 Airlines with Bird Strikes')
plt.xlabel('Airline')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Airports with most incidents of bird strikes  Top 50
top_50_airports = bird_strikes_data['Airport: Name'].value_counts().nlargest(50)

# Plotting Top 50 Airports
plt.figure(figsize=(12, 8))
top_50_airports.plot(kind='barh', color='green')
plt.title('Top 50 Airports with Bird Strikes')
plt.xlabel('Number of Bird Strikes')
plt.ylabel('Airport')
plt.tight_layout()
plt.show()

# Yearly Cost Incurred due to Bird Strikes
yearly_cost = bird_strikes_data.groupby('Year')['Cost: Total $'].sum()

# Plotting Yearly Cost
plt.figure(figsize=(10, 6))
yearly_cost.plot(kind='bar', color='purple')
plt.title('Yearly Cost Incurred due to Bird Strikes')
plt.xlabel('Year')
plt.ylabel('Cost ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# When do most bird strikes occur?
phase_of_flight_counts = bird_strikes_data['When: Phase of flight'].value_counts()

# Plotting Phase of Flight
plt.figure(figsize=(10, 6))
phase_of_flight_counts.plot(kind='bar', color='brown')
plt.title('When do most Bird Strikes Occur?')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Altitude of airplanes at the time of strike
altitude_counts = bird_strikes_data['Altitude bin'].value_counts()

# Plotting Altitude of Strikes
plt.figure(figsize=(10, 6))
altitude_counts.plot(kind='bar', color='pink')
plt.title('Altitude of Airplanes at the Time of Strike')
plt.xlabel('Altitude')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
